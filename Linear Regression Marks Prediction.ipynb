## Marks Prediction System

Goal ->

We use Linear Regression modal to predict a continuous value of Marks based on Hours of sleeping.

Model : LINEAR REGRESSION

Equation:
        y = mx + c

Where:

y = predicted value,
x = input feature,
c = intercept,
m = slope or coefficient

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
#sample data
X = np.array([[1],[2],[5],[3],[4]])
y = np.array([20,40,60,50,60])
```

```python
#modal
modal = LinearRegression()
modal.fit(X,y)
```

```python
#prediction
y_pred = modal.predict(X)
y_pred
```

```python
#coefficent
print("Slope(m) : ",modal.coef_[0])
print("Intercept(c) ",modal.intercept_)
```

```python
#plot
plt.scatter(X,y,color='blue')
plt.plot(X,y_pred,color='red')
plt.title("Linear Regression")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.show()
```
